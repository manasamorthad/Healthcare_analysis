```python
#1: Creating Arrays from Dataset Columns
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("archive/healthcare_dataset.csv")

# Creating arrays from dataset columns
ages = df['Age'].values
billing = df['Billing Amount'].values

# Display the arrays
print("Ages array:\n", ages)
print("\nBilling Amount array:\n", billing)
```

```python
# Array Indexing and Slicing Examples
# Get first 10 Age values
print("First 10 Age values:")
print(ages[:10])

# Get last 5 Billing Amount values
print("\nLast 5 Billing Amount values:")
print(billing[-5:])

```

```python
# Reshaping Arrays
# Reshape ages into 2 rows and n/2 columns (example: first 10)
reshaped_ages = ages[:10].reshape(2, 5)
print("\nReshaped Age Array (2x5):")
print(reshaped_ages)
```

```python
#Concatenation and Splitting
# Concatenate first 5 ages and billings (as 2D array)
combined = np.column_stack((ages[:5], billing[:5]))
print("\nConcatenated Ages and Billings:")
print(combined)

# Split back into two arrays
split_ages, split_billings = np.hsplit(combined, 2)
print("\nSplit Ages:")
print(split_ages.flatten())
print("\nSplit Billings:")
print(split_billings.flatten())
```

```python
# NumPy universal functions
# Square root of all billing values (safely ignoring negatives)
print("Square root function:")
print(np.sqrt(np.clip(billing, 0, None)))

# Multiply billing by 2
print("\nMultiplication function:")
print(np.multiply(billing, 2))

# Divide ages by 2
print("\nDivision function:")
print(np.divide(ages, 2))

# Exponential of billing values (safe using clip)
print("\nExponential function (safe):")
print(np.exp(np.clip(billing[:5], 0, 700)))  # limit to avoid overflow

# Logarithm of billing values (only valid ones)
print("\nLogarithmic function (valid only):")
valid_billing = billing[billing > 0]
print(np.log(valid_billing))

# Convert ages to Radians and Compute Sine
print("\nSin function (from Age):")
print(np.sin(np.radians(ages)))

# Cosine of billing values
print("\nCos function (from Billing Amount):")
print(np.cos(np.radians(billing)))

# Absolute values (if any negatives in billing)
print("\nAbsolute function:")
print(np.abs(billing))

# Rounding Billing Amount values to nearest integer
print("\nRounding function:")
print(np.round(billing))

# Check which patients have Billing > 30000
print("\nGreater than operator (Billing > 30000):")
print(np.greater(billing, 30000))

# Check which patients have Age < 40
print("\nLess than operator (Age < 40):")
print(np.less(ages, 40))

# Find patients whose Age is exactly 50
print("\nEqual operator (Age == 50):")
print(np.equal(ages, 50))
```

```python
#Aggregation Functions
# Sum aggregation
total_age = np.sum(ages)

# Mean (Average)
mean_billing = np.mean(billing)

# Median
median_age = np.median(ages) 

# Standard Deviation
std_billing = np.std(billing) 

# Variance
var_age = np.var(ages) 

# Minimum and Maximum
min_billing = np.min(billing)
max_billing = np.max(billing)  

print("\nTotal age sum: ", total_age)
print("\nMean billing amount: ", mean_billing)
print("\nMedian age: ", median_age)
print("\nStandard Deviation in Billing: ", std_billing)
print("\nVariance in Age: ", var_age)
print("\nLowest billing: ", min_billing)
print("\nHighest billing: ", max_billing)

# Cumulative Sum
print("\nCumulative sum of Ages: ")
print(np.cumsum(ages))  # Running total of Ages
```

```python
#sorting
# Sorting patients by Billing Amount in descending order
print("Patients sorted by Billing Amount (High to Low):")
df_sorted_billing = df.sort_values("Billing Amount", ascending=False)
print(df_sorted_billing)

print("\n------------------------------------------\n")

# Sorting patients by Age in ascending order
print("Patients sorted by Age (Low to High):")
df_sorted_age = df.sort_values("Age", ascending=True)
print(df_sorted_age)
```

```python
#Fancy Indexing & Partial Sorting
# Convert columns to arrays
billing_array = np.array(df["Billing Amount"])
ages_array = np.array(df["Age"])

# Get the indices of sorted Billing Amount values
sorted_indices = np.argsort(billing_array)
print("Indices that would sort Billing Amounts:")
print(sorted_indices)

# Partition at index to find top 5 highest Billing Amounts
top_5_billing = np.partition(billing_array, -5)[-5:]
print("\nTop 5 highest Billing Amounts:")
print(top_5_billing)
```

```python
#Joins
import pandas as pd

# Load your original healthcare dataset
df = pd.read_csv("archive/healthcare_dataset.csv")

# Sample extra dataset for joining
extra_info = pd.DataFrame({
    "Name": ["Bobby JacksOn", "LesLie TErRy", "Non Existent", "DaNnY sMitH"],
    "Follow-up Needed": ["Yes", "No", "Yes", "No"]
})

# 1. INNER JOIN
inner_join = df.merge(extra_info, on="Name", how="inner")
print("\n🔹 Inner Join:")
print(inner_join)

# 2. LEFT JOIN
left_join = df.merge(extra_info, on="Name", how="left")
print("\n🔹 Left Join:")
print(left_join)

# 3. RIGHT JOIN
right_join = df.merge(extra_info, on="Name", how="right")
print("\n🔹 Right Join:")
print(right_join)

# 4. OUTER JOIN
outer_join = df.merge(extra_info, on="Name", how="outer")
print("\n🔹 Outer Join:")
print(outer_join)
```

```python
#Aggregation and Grouping
# Group by Medical Condition and aggregate Billing Amount
grouped_billing = df.groupby("Medical Condition")["Billing Amount"].agg(
    ["sum", "count", "mean", "median", "min", "max"]
)
print("\n🔹 Billing Amount Statistics by Medical Condition:")
print(grouped_billing)

# Group by Insurance Provider and aggregate Age
grouped_age = df.groupby("Insurance Provider")["Age"].agg(
    ["sum", "count", "mean", "median", "min", "max"]
)
print("\n🔹 Age Statistics by Insurance Provider:")
print(grouped_age)

```

```python
#pivot tables
#Pivot Table 1: Average Billing per Medical Condition
pivot1 = pd.pivot_table(df, index="Medical Condition", values="Billing Amount", aggfunc="mean")
print("Pivot Table - Average Billing by Medical Condition:")
print(pivot1)
grouped1 = df.groupby("Medical Condition")["Billing Amount"].mean()
print("GroupBy - Average Billing by Medical Condition:")
print(grouped1)
```

```python
#Pivot Table 2: Maximum Billing Amount per Gender and Insurance Provider
pivot2 = pd.pivot_table(df, index="Gender", columns="Insurance Provider", values="Billing Amount", aggfunc="max")
print("\nPivot Table - Max Billing by Gender & Insurance Provider:")
print(pivot2)
grouped2 = df.groupby(["Gender", "Insurance Provider"])["Billing Amount"].max().unstack()
print("GroupBy - Max Billing by Gender & Insurance Provider:")
print(grouped2)
```

```python
# Pivot Table with Multiple Aggregations (Min, Max, Mean Billing Amount)
pivot_table = df.pivot_table(
    values="Billing Amount",
    index="Medical Condition",
    aggfunc=["min", "max", "mean"]
)
print(pivot_table)
```

```python
#LINEPLOT
import matplotlib.pyplot as plt

# Sample the first 10 rows
sample_df = df.head(10)

# Line Plot: Age vs Billing Amount (First 10 entries)
plt.plot(sample_df["Age"], sample_df["Billing Amount"], color="blue", marker="o")
plt.title("Age vs Billing Amount (First 10 Patients)")
plt.xlabel("Age")
plt.ylabel("Billing Amount")
plt.grid(True)
plt.show()
```

```python
#SCATTERPLOT
plt.scatter(sample_df["Age"], sample_df["Billing Amount"], color="green")
plt.title("Scatter: Age vs Billing Amount (First 10 Patients)")
plt.xlabel("Age")
plt.ylabel("Billing Amount")
plt.show()
```

```python
#BAR GRAPH- Bar Chart – Billing Amount per Patient
import matplotlib.pyplot as plt
sample_df = df.head(10)

plt.figure(figsize=(10, 5))
plt.bar(sample_df["Name"], sample_df["Billing Amount"], color="purple")
plt.xticks(rotation=45)
plt.title("Billing Amount per Patient")
plt.xlabel("Patient Name")
plt.ylabel("Billing Amount")
plt.tight_layout()
plt.show()
```

```python
#HISTOGRAM-AGE DISTRIBUTION
plt.hist(sample_df["Age"], bins=5, color="orange", edgecolor="black")
plt.title("Age Distribution (First 10 Patients)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Pie Chart – Admission Types Distribution
plt.figure(figsize=(6,6))
admission_counts = sample_df["Admission Type"].value_counts()
plt.pie(admission_counts, labels=admission_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Admission Type Distribution")
plt.show()
```
